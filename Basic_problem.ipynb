```python
def factorial(n):
    """
    Calculates the factorial of a non-negative integer.
    """
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Get input from the user
num_str = input("Enter a non-negative integer to compute its factorial: ")

try:
    num = int(num_str)
    result = factorial(num)
    print(f"The factorial of {num} is {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
num_str = input("Enter a number to display its multiplication table: ")

try:
    num = int(num_str)
    print(f"\nMultiplication Table for {num}:")
    for i in range(1, 11): # Table from 1 to 10
        print(f"{num} x {i} = {num * i}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
num_str = input("Enter an integer to check if it's odd or even: ")

try:
    num = int(num_str)
    if num % 2 == 0:
        print(f"The number {num} is Even.")
    else:
        print(f"The number {num} is Odd.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
num_str = input("Enter a number to check if it's a palindrome: ")

try:
    # Convert the input string to an integer first to validate it's a number
    num = int(num_str)

    # Convert the integer back to a string for palindrome check
    # This handles cases like 010 being treated as 10
    s_num = str(num)

    # Reverse the string
    reversed_s_num = s_num[::-1]

    # Compare the original string with the reversed string
    if s_num == reversed_s_num:
        print(f"The number {num} is a palindrome.")
    else:
        print(f"The number {num} is not a palindrome.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
def run_expert_system():
    print("Welcome to the Drink Recommendation Expert System!")
    print("Please answer a few questions to get your perfect drink.\n")

    # 1. Define rules
    # Each rule is a dictionary with 'conditions' (a dict of criteria) and 'recommendation'
    rules = [
        {
            "conditions": {"caffeine_needed": "yes", "hot_or_cold": "hot", "sweet_preference": "no"},
            "recommendation": "Black Coffee"
        },
        {
            "conditions": {"caffeine_needed": "yes", "hot_or_cold": "hot", "sweet_preference": "yes"},
            "recommendation": "Latte or Cappuccino"
        },
        {
            "conditions": {"caffeine_needed": "yes", "hot_or_cold": "cold", "sweet_preference": "yes"},
            "recommendation": "Iced Coffee or Frappuccino"
        },
        {
            "conditions": {"caffeine_needed": "no", "hot_or_cold": "hot", "sweet_preference": "no"},
            "recommendation": "Herbal Tea"
        },
        {
            "conditions": {"caffeine_needed": "no", "hot_or_cold": "hot", "sweet_preference": "yes"},
            "recommendation": "Hot Chocolate or Sweet Tea"
        },
        {
            "conditions": {"caffeine_needed": "no", "hot_or_cold": "cold", "sweet_preference": "no"},
            "recommendation": "Water or Unsweetened Iced Tea"
        },
        {
            "conditions": {"caffeine_needed": "no", "hot_or_cold": "cold", "sweet_preference": "yes"},
            "recommendation": "Fruit Juice or Lemonade"
        }
    ]

    # 2. Formulate questions corresponding to the conditions in rules
    questions = {
        "caffeine_needed": "Do you need caffeine? (yes/no): ",
        "hot_or_cold": "Do you prefer a hot or cold drink? (hot/cold): ",
        "sweet_preference": "Do you prefer a sweet drink? (yes/no): "
    }

    user_responses = {}

    # 3. Implement mechanism to ask questions and store responses
    for key, question_text in questions.items():
        while True:
            answer = input(question_text).lower().strip()
            if key == "caffeine_needed" and answer in ["yes", "no"]:
                user_responses[key] = answer
                break
            elif key == "hot_or_cold" and answer in ["hot", "cold"]:
                user_responses[key] = answer
                break
            elif key == "sweet_preference" and answer in ["yes", "no"]:
                user_responses[key] = answer
                break
            else:
                print("Invalid input. Please try again.")

    # 4. Write logic to evaluate user responses against rules
    decision = "I'm sorry, I cannot make a specific recommendation based on your preferences. Maybe try something new!"

    for rule in rules:
        match = True
        for condition_key, condition_value in rule["conditions"].items():
            if user_responses.get(condition_key) != condition_value:
                match = False
                break
        if match:
            decision = rule["recommendation"]
            break # Found a matching rule, exit loop

    # 5. Display the decision
    print("\n--- Your Recommendation ---")
    print(f"Based on your preferences, we recommend: {decision}")
    print("-------------------------")

# Run the expert system
run_expert_system()
```
